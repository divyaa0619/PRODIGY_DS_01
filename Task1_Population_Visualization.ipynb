```python
import pandas as pd

df = pd.read_csv('worldpopulationdata.csv')
df.head()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df1 = df[['Country Name','2008']]

df1 = df1.dropna()
df1['2008] = pd.to_numeric(df1['2021'], errors='coerce')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


df1_fixed = df[['Country Name', '2021']].copy()
df1_fixed = df1_fixed.dropna()
df1_fixed['2021'] = pd.to_numeric(df1_fixed['2021'], errors='coerce')

plt.figure()
sns.histplot(df1_fixed['2021'], bins=30)
plt.title("Population Distribution (2021)")
plt.xlabel("population")
plt.ylabel("Frequency")
plt.show()
```

```python
top10 = df1.sort_values(by='2021', ascending=False).head(10)

plt.figure()
plt.bar(top10['Country Name'], top10['2021'])
plt.title("Top 10 Most Populated Countries (2021)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.show()
```
